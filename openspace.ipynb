```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import io, measure
```

```python
img = io.imread('sample.png')
plt.imshow(img)
```

```python
print(img.shape)
```

```python
# Convert RGBA to RGB
if img.shape[-1] == 4:  # Check if the image has 4 channels
    img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
mask = cv2.inRange(hsv, (20, 20, 20), (50, 255, 255))
```

```python
plt.imshow(img)
```

```python
from scipy import ndimage as ndi
```

```python
closed_mask = ndi.binary_closing(mask, structure=np.ones((7, 7)))
```

```python
plt.imshow(closed_mask)
```

```python
label_image = measure.label(closed_mask)
```

```python
plt.imshow(label_image)
```

```python
from skimage.color import label2rgb
```

```python
print(img.shape)
```

```python
# Convert RGBA to RGB
if img.shape[-1] == 4:  # Check if the image has 4 channels
    img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)

# Generate the label overlay
image_label_overlay = label2rgb(label_image, image=img)

# Display the result
plt.imshow(image_label_overlay)
plt.show()
```
